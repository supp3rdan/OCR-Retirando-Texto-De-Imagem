```python
from PIL import Image
import pytesseract
```

```python
Image.open('teste.JPG')
```

```python
print(pytesseract.image_to_string( Image.open('teste.JPG'), lang='por'))
```

```python
texto = pytesseract.image_to_string( Image.open('teste.JPG'), lang='por')
texto2 = texto.replace('\n',' ').replace('-','').replace('\x0c','')
arquivo = open('novo-arquivo.txt', 'w')
arquivo.write(texto2)
arquivo.close()
```
